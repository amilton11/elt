```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from datetime import datetime
```

```python
# pull in DJIA data csv
file = "Data/Dow Jones Industrial Average Historical Data.csv"

djia_df = pd.read_csv(file)
djia_df.head()
```

```python
# pull in store data csv
file2 = "Data/train.csv"

store_df = pd.read_csv(file2)
store_df.head()
```

```python
#put date info in date time format
djia_df['Date']= pd.to_datetime(djia_df["Date"])
djia_df.head()
```
